```python
import pandas as pd
df1=pd.read_csv('/Users/cra/Desktop/completevirat.csv',index_col=False)
df2=pd.read_csv('VIRAT FOR CR.csv',index_col=False)
df3=pd.read_excel('test virat.xlsx')
```

# PREPROCESSING

```python
data=df1.iloc[:,1:]
data['format']='odi'
data.head()
data=data.drop(['Ground'],axis=1)
data.head()
data.rename(columns = {'runs':'Runs'}, inplace = True)
data.rename(columns = {'Opp':'Opposition'}, inplace = True)
data['Runs'] = pd.to_numeric(data['Runs'], errors='coerce')
data.dropna(inplace=True)
# data.reset_index(drop=True, inplace=True)
data
```

```python
df2
columns_to_select = ['Runs', 'Opposition', 'Ground','Inns']
datato=df2[columns_to_select]
datato['format']='t20'
datato=datato.drop(['Ground'],axis=1)
datato["Opposition"]=datato["Opposition"].str.replace("v","")
datato["Runs"]=datato["Runs"].str.replace("*","")

datato['Runs'] = pd.to_numeric(datato['Runs'], errors='coerce')
datato.dropna(inplace=True)
datato
```

```python
frames = [data, datato]
  
result = pd.concat(frames)
result
```

```python
columns_to_select = ['Score', 'Opposition','Inns']
test=df3[columns_to_select]
test['format']='test'
test
def fun(num):
  
    if num==3:
        return '1'
    elif num==4:
        return '2'
    elif num==1:
        return '1'
    else:
        return '2'
test['Inns']=test['Inns'].apply(fun)
test
    
    
```

```python
test = test[test["Score"].str.contains("-") == False]
test["Opposition"]=test["Opposition"].str.replace("v","")
test["Score"]=test["Score"].str.replace("*","")
test.rename(columns = {'Score':'Runs'}, inplace = True)
test['Runs'] = pd.to_numeric(test['Runs'], errors='coerce')
test.dropna(inplace=True)
test
```

```python
test.info()
```

```python
frames = [result,test]
  
res = pd.concat(frames)
res.reset_index(inplace=True,drop=True)
res
```

```python
res['Opposition']=res['Opposition'].str.replace(' ','')
def fun(num):
  
    if num<30:
        return 0
  
    elif num>= 30 and num<50:
        return 1
  
    else:
        return 2
res['Runs']=res['Runs'].apply(fun)
res
```

## One hot encoding done to handle categorical datatypes

```python
x=res.iloc[:,1:]
y=res.iloc[:,:1]


state = pd.get_dummies(res['Opposition'])
x = x.join(state)
x = x.drop(['Opposition'],axis=1)

state2 = pd.get_dummies(res['format'])
x = x.join(state2)
x = x.drop(['format'],axis=1)

state3 = pd.get_dummies(res['Inns'])
x = x.join(state3)
x = x.drop(['Inns'],axis=1)
x
```

## Checking for multicolinearity between independent columns

```python
vif = pd.DataFrame()
from statsmodels.stats.outliers_influence import \
variance_inflation_factor

vif['variable']=x.columns

vif['vif']= [variance_inflation_factor(x.values,i)
             for i in range(x.shape[1])]

vif
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,
                                                random_state=80)

```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion='gini')
tree.fit(x_train,y_train)
```

```python
y_pred = tree.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

## Using random forest and grid search to obtain the best model

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100,
                            criterion='gini')
```

```python
rfc.fit(x_train,y_train)
from sklearn.metrics import accuracy_score
y_pred = rfc.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param = {'n_estimators':[25,50,75,100,150,200],
        'criterion':['gini','entropy'],
        'bootstrap':[True, False]}
cls_grid = RandomForestClassifier()
gr = GridSearchCV(cls_grid, param, cv=10, n_jobs=-1)

gr.fit(x_train, y_train)
print(gr.best_params_)
gr.best_score_
```

# Neural Network with multiple layers to obtain the best result

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Dense(units=1024,activation='relu',input_shape=(20,)))
model.add(Dense(units=512,activation='relu'))
model.add(Dense(units=256,activation='relu'))
model.add(Dense(units=128,activation='relu'))
model.add(Dense(units=64,activation='relu'))
model.add(Dense(units=3, activation='softmax'))

model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=['accuracy'])


model.fit(x,y , epochs=100, batch_size=8)


loss, accuracy = model.evaluate(x_test, y_test)
print("Testing accuracy:", accuracy)
```
